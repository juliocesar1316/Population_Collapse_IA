```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Populaçao total
total_population = pd.read_csv("./dataset/total_population.csv")
display(total_population.head())
```

```python
# Transformar de wide para long format
total_population_long = pd.melt(total_population, id_vars=["economy"], var_name="year", value_name="total_population")
# Remover o prefixo 'YR' da coluna 'Ano'
total_population_long["year"] = total_population_long["year"].str.replace("YR", "").astype(int)

display(total_population_long)
```

```python
# # Filtrar por uma "economy" específica, como 'ABW', para evitar sobrecarga visual
# total_population_br = total_population_long[total_population_long["economy"] == "BRA"]


# Plotar o gráfico com todas as economias ao longo do tempo
plt.figure(figsize=(20, 10))
sns.lineplot(x="year", y="total_population", hue="economy", data=total_population_long, palette="viridis")
plt.title('Total Population by Economy', fontsize=16, weight='bold')
plt.xlabel('Ano', fontsize=14, weight='bold')
plt.ylabel('Populacao', fontsize=14, weight='bold')
plt.xticks(rotation=45)
plt.show()
```
